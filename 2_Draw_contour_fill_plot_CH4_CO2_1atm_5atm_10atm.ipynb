```python
# Change of target gas with temperature and raw material ratio, 0.1Mpa

import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file
df = pd.read_excel('CH4-CO2-1atm.xlsx')

# View the first few lines of data
print(df.head())

# Extract the required columns
# The column name can be modified, but it must match the column name in table execl
T = df['T']
CO2_CH4 = df['CO2/CH4']  
H2_CO = df['H2+CO']

# Set the global font size and thickness
plt.rcParams['font.size'] = 15  
plt.rcParams['font.weight'] = 'semibold'  

# Draw a contour fill plot
plt.figure(figsize=(8, 6))
contourf = plt.tricontourf(T, CO2_CH4, H2_CO, levels=500, cmap='coolwarm', alpha=1, vmin=0, vmax=0.93)  # Adjust the color mapping range
cbar = plt.colorbar(contourf)
cbar.set_label('H2+CO', fontsize=15, weight='semibold')  # Set colorbar font size and weight
plt.xlabel('T/K', fontsize=15, weight='semibold')
plt.ylabel('CO2/CH4', fontsize=15, weight='semibold')
#plt.title('percentage of purpose gas at 1atm', fontsize=15, weight='semibold')

# Add contour lines with H2+CO values of 0.5, 0.7 and 0.9
levels = [0.4,0.5,0.7,0.8,0.9]
contour = plt.tricontour(T, CO2_CH4, H2_CO, levels=levels, colors='black', linestyles='dashed', linewidths=3)  # Set contour thickness
plt.clabel(contour, inline=True, fontsize=18, fmt={level: str(level) for level in levels})

plt.grid(False)  # Do not display gridlines
plt.show()
```
